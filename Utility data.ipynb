```python
# getting the id for utilities that do not have an unexpired rate plan.
Util_all=df["eiaid"].unique().tolist()
Util_nonex=df[df["enddate"].isna()]["eiaid"].tolist()
Util_diff= list(set(Util_all) - set(Util_nonex))
Util_diff
```

```python
#joining database of unexpired plans with the plans for utilies that have older(expired) rate plans
df_a=df[df['enddate'].isna()]
df_b=df[df['eiaid'].isin(Util_diff)]
df_full=df_a.append(df_b, ignore_index=True)
df_full.head(10)
```
